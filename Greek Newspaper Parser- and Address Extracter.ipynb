```python
from docx import *
#to install docx, run "pip install python-docx"
#if that doesn't work, run "conda install -c conda-forge python-docx"
import numpy as np
import pandas as pd
import re

""" 
Label (underlined)
Subtitle:
Duration:
Publisher:
Director:
Address:
Printing house:
"""

document = Document('Greek Newspapers of Constantinople List.docx')
```

```python
# Print document, for visibility
print('--------------------------')
for line, p in enumerate(document.paragraphs):
    print("Paragraph {}:".format(line))
    print(p.text)
    print()
print('--------------------------')
```

```python
# This cell loads all the data into a dataframe

cols = pd.Series(["Label", "Subtitle", "Duration", "Publisher", "Director", "Address", "Printing house"])
df = pd.DataFrame(columns = cols)
dftemp = pd.DataFrame(columns = cols) # temporary dataframe

true_doc_start = 3 # Since the first 2 "paragraphs" are junk data, we want to start parsing at line 3 
                   # We start at line 3 because the format is blank line - label - rest of columns

for line, p in enumerate(document.paragraphs):
    if line < true_doc_start: # Ignore the first couple of lines of junk
        continue;
        
    if p.text == "": # If its a newline, we append the old dftemp and then reset it
        if line > 4: 
            df = pd.concat([df, dftemp], ignore_index=True)
        dftemp = pd.DataFrame(columns = cols)
    else:
        m = re.search(":", p.text) #see what the key:value format is
        if m == None: # 
            dftemp["Label"] = [p.text] # if t
        else:
            curr = p.text.split(":")
            dftemp[curr[0]] = [curr[1]]

df
```

```python
#loads dataframe as is into EXCEL document
df.to_excel("UNFILTERED-Greek Newspapers of Constantinople List.xlsx")  
```

```python
# This is a the dataframe, with all the unwanted columns simply cut out.

trimmed_df = df.loc[:, "Label": "Printing house"]
trimmed_df

trimmed_df.to_excel("LOSSY_CUT-Greek Newspapers of Constantinople List.xlsx")  
```

```python
## This is the new cell that extracts the addresses into a new column, and then converts 

cols = pd.Series(["Label", "Subtitle", "Duration", "Publisher", "Director", "Printing house"])
df_add = pd.DataFrame(columns = cols)


for index, row in trimmed_df.iterrows():
    curr_add = row["Address"]
    curr_row = row.drop(labels = "Address")
    if type(curr_add) == str:
        m = re.search(";", curr_add)
        if m == None:
            curr_row["Address 1"] = curr_add
        else:
            curr_list = curr_add.split(";")
            for i in range(len(curr_list)):
                add_str = "Address " + str(1 + i)
                curr_row[add_str] = curr_list[i]
    
    
    df_add = df_add.append(curr_row)
    
df_add.to_excel("ADDRESS_LIST-Greek Newspapers of Constantinople List.xlsx")  
```

```python
## This is the new cell that extracts the addresses into a new column, and then converts 
import geopandas as gpy

cols = pd.Series(["Label", "Subtitle", "Duration", "Publisher", "Director", "Printing house"])
geo_df = pd.DataFrame(columns = cols)


for index, row in trimmed_df.iterrows():
    curr_add = row["Address"]
    curr_row = row.drop(labels = "Address")
    if type(curr_add) == str:
        m = re.search(";", curr_add)
        if m == None:
            curr_row["Address 1"] =  gpy.geocode(strings =[curr_add])
        else:
            curr_list = curr_add.split(";")
            for i in range(len(curr_list)):
                add_str = "Address " + str(1 + i)
                curr_row[add_str] = curr_list[i]
    
    
    geo_df = geo_df.append(curr_row)
    
geo_df.to_excel("GEOCODE_LIST-Greek Newspapers of Constantinople List.xlsx")  
```
